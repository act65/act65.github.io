```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def e1_benthamite(u1, u2, u3):
  """Calculates the first elementary symmetric polynomial (Benthamite)."""
  return u1 + u2 + u3

def e2(u1, u2, u3):
  """Calculates the second elementary symmetric polynomial."""
  return (u1 * u2) + (u1 * u3) + (u2 * u3)

def e3_nash(u1, u2, u3):
  """Calculates the third elementary symmetric polynomial (Nash)."""
  return u1 * u2 * u3

u1 = 1.0
u2 = 1.0
u3 = -2.0

"""Generates a 3D scatter plot to visualize the utility functions."""
# Generate a grid of utility values
n_points = 10
u = np.linspace(0.1, 1.0, n_points)
u1, u2, u3 = np.meshgrid(u, u, u)
u1 = u1.flatten()
u2 = u2.flatten()
u3 = u3.flatten()

# Calculate the utility functions for each point
val_e1 = e1_benthamite(u1, u2, u3)
val_e2 = e2(u1, u2, u3)
val_e3 = e3_nash(u1, u2, u3)

# Create the 3D plot
fig = plt.figure(figsize=(18, 6))
fig.suptitle('Visualizing e1, e2, and e3 in Utility Space (u1, u2, u3)', fontsize=16)

# Plot for e1
ax1 = fig.add_subplot(131, projection='3d')
sc1 = ax1.scatter(u1, u2, u3, c=val_e1, cmap='viridis', s=50)
ax1.set_title('e1 (Benthamite)')
ax1.set_xlabel('u1')
ax1.set_ylabel('u2')
ax1.set_zlabel('u3')
fig.colorbar(sc1, ax=ax1, shrink=0.5, aspect=10, label='e1 Value')

# Plot for e2
ax2 = fig.add_subplot(132, projection='3d')
sc2 = ax2.scatter(u1, u2, u3, c=val_e2, cmap='plasma', s=50)
ax2.set_title('e2')
ax2.set_xlabel('u1')
ax2.set_ylabel('u2')
ax2.set_zlabel('u3')
fig.colorbar(sc2, ax=ax2, shrink=0.5, aspect=10, label='e2 Value')

# Plot for e3
ax3 = fig.add_subplot(133, projection='3d')
sc3 = ax3.scatter(u1, u2, u3, c=val_e3, cmap='inferno', s=50)
ax3.set_title('e3 (Nash)')
ax3.set_xlabel('u1')
ax3.set_ylabel('u2')
ax3.set_zlabel('u3')
fig.colorbar(sc3, ax=ax3, shrink=0.5, aspect=10, label='e3 Value')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import itertools

def ec2v2(*us):
    # print(us)
    if len(us) == 2:
        return us[0] * us[1]
    else:
        return sum([ec2v2(*x) for x in itertools.permutations(us, len(us)-1)])/len(us)
```

```python
us = [0.5, 0.3, 0.7, 1.5]

xs = np.linspace(0.0, 1.0, 20)
y = np.zeros((3, 20))

for i, x in enumerate(xs):
    us_ = us + [float(x)]
    y[0, i] = np.sum(us_)/len(us)
    y[1, i] = ec2v2(*us_)
    y[2, i] = np.prod(us_)
```
